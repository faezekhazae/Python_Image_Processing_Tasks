```python
import cv2  # Import the OpenCV library for computer vision tasks
# وارد کردن کتابخانه OpenCV برای انجام کارهای بینایی ماشین

import matplotlib.pyplot as plt  # Import matplotlib for plotting images and graphs
# وارد کردن کتابخانه matplotlib برای رسم تصاویر و نمودارها

import numpy as np  # Import NumPy for numerical operations and array handling
# وارد کردن کتابخانه NumPy برای انجام عملیات عددی و مدیریت آرایه‌ها
```

```python
image = cv2.imread('./lena_gray_256.tif', 0)  # Read the image in grayscale mode (0 means grayscale)
# خواندن تصویر با حالت خاکستری (۰ به معنی خاکستری است)

# window_name = 'img'  # (Commented out) This could be used for naming a window if using cv2.imshow
# (غیرفعال شده) این می‌توانست برای نام‌گذاری پنجره در صورت استفاده از cv2.imshow استفاده شود

plt.imshow(image, cmap="gray")  # Display the image using matplotlib with a grayscale colormap
# نمایش تصویر با matplotlib با استفاده از نقشه رنگ خاکستری

plt.axis('off')  # Turn off the axes for a cleaner display
# خاموش کردن محور‌ها برای نمایش مرتب‌تر

plt.show()  # Show the plotted image
# نمایش تصویر رسم شده
```

```python
histogram = cv2.calcHist([image], [0], None, [256], [0, 256])  
# Calculate the histogram of the grayscale image
# محاسبه هیستوگرام تصویر خاکستری
# [image] -> list of images to process
# [0] -> channel index (0 for grayscale)
# None -> no mask is used
# [256] -> number of bins
# [0,256] -> intensity range

plt.plot(histogram, color='k')  # Plot the histogram in black color ('k')
# رسم هیستوگرام با رنگ مشکی

plt.show()  # Display the histogram plot
# نمایش نمودار هیستوگرام
```

```python
img = cv2.imread('bird.jpg')  # Read the color image using OpenCV (BGR format)
# خواندن تصویر رنگی با استفاده از OpenCV (فرمت BGR)

plt.imshow(img)  # Display the image using matplotlib
# نمایش تصویر با استفاده از matplotlib

plt.axis('off')  # Turn off axes for a cleaner display
# خاموش کردن محور‌ها برای نمایش مرتب‌تر

plt.show()  # Show the plotted image
# نمایش تصویر رسم شده


for i, col in enumerate(['b', 'g', 'r']):  # Loop over the BGR channels
    # حلقه روی کانال‌های BGR
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])  
    # Calculate histogram for the current channel
    # محاسبه هیستوگرام برای کانال فعلی
    plt.plot(hist, color=col)  # Plot the histogram using the corresponding color
    # رسم هیستوگرام با رنگ متناظر (B->blue, G->green, R->red)
    plt.xlim([0, 256])  # Set the x-axis limits from 0 to 255
    # تنظیم محدوده محور x از ۰ تا ۲۵۵
    plt.show()  # Display the histogram plot
    # نمایش نمودار هیستوگرام
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  
# Convert the color image to grayscale using OpenCV
# تبدیل تصویر رنگی به خاکستری با استفاده از OpenCV

gray_img_eqhist = cv2.equalizeHist(gray_img)  
# Apply histogram equalization to enhance the contrast of the grayscale image
# اعمال متوازن‌سازی هیستوگرام برای افزایش کنتراست تصویر خاکستری
```

```python
plt.imshow(gray_img, cmap="gray")  # Display the grayscale image using a gray colormap
# نمایش تصویر خاکستری با استفاده از نقشه رنگ خاکستری

plt.axis('off')  # Turn off axes for a cleaner display
# خاموش کردن محور‌ها برای نمایش مرتب‌تر

plt.show()  # Show the plotted grayscale image
# نمایش تصویر خاکستری رسم شده

```

```python
plt.plot(gray_img_eqhist, color='k')  
# Attempt to plot the equalized grayscale image as a line plot in black ('k')
# تلاش برای رسم تصویر خاکستری متوازن شده به صورت نمودار خطی با رنگ مشکی ('k')
# ⚠️ Note: This will not display the image correctly; images should use plt.imshow(), not plt.plot()

plt.show()  # Display the plot
# نمایش نمودار
```

```python
ret = cv2.threshold(gray_img, 80, 256, cv2.THRESH_BINARY)  
# Apply a binary threshold to the grayscale image
# اعمال آستانه‌گذاری باینری روی تصویر خاکستری
# Pixels with intensity > 80 will be set to 256, others to 0
# پیکسل‌هایی با شدت > ۸۰ برابر ۲۵۶ می‌شوند و بقیه برابر ۰
# cv2.THRESH_BINARY specifies binary thresholding
# cv2.THRESH_BINARY مشخص می‌کند که آستانه‌گذاری به صورت باینری انجام شود
```

```python
flip = cv2.flip(img, 0)  
# Flip the image vertically (0 means vertical flip)
# برگرداندن تصویر به صورت عمودی (۰ یعنی عمودی)

plt.imshow(flip)  # Display the flipped image using matplotlib
# نمایش تصویر برگردانده شده با استفاده از matplotlib

plt.axis('off')  # Turn off axes for a cleaner display
# خاموش کردن محور‌ها برای نمایش مرتب‌تر

plt.show()  # Show the plotted flipped image
# نمایش تصویر برگردانده شده
```

```python
blurImg = cv2.blur(img, (20, 20))  
# Apply a simple averaging blur to the image with a 20x20 kernel
# اعمال محوشدگی متوسط ساده روی تصویر با کرنل ۲۰ در ۲۰

plt.imshow(blurImg)  # Display the blurred image using matplotlib
# نمایش تصویر محوشده با استفاده از matplotlib

plt.axis('off')  # Turn off axes for a cleaner display
# خاموش کردن محور‌ها برای نمایش مرتب‌تر

plt.show()  # Show the plotted blurred image
# نمایش تصویر محوشده رسم شده
```

```python
gausBlur = cv2.GaussianBlur(img, (5, 5), 0)  
# Apply a Gaussian blur to the image with a 5x5 kernel and sigma=0
# اعمال محوشدگی گاوسی روی تصویر با کرنل ۵ در ۵ و سیگما=۰
# Gaussian blur smooths the image while preserving edges better than average blur
# محوشدگی گاوسی تصویر را نرم می‌کند و نسبت به محوشدگی متوسط لبه‌ها را بهتر حفظ می‌کند

plt.imshow(gausBlur)  # Display the Gaussian blurred image using matplotlib
# نمایش تصویر محوشده گاوسی با استفاده از matplotlib

plt.axis('off')  # Turn off axes for a cleaner display
# خاموش کردن محور‌ها برای نمایش مرتب‌تر

plt.show()  # Show the plotted Gaussian blurred image
# نمایش تصویر محوشده گاوسی رسم شده

```

```python
medBlur = cv2.medianBlur(img, 5)  
# Apply a median blur to the image with a kernel size of 5
# اعمال محوشدگی میانه روی تصویر با اندازه کرنل ۵
# Median blur helps reduce salt-and-pepper noise while preserving edges
# محوشدگی میانه به کاهش نویز نمک و فلفل کمک می‌کند و لبه‌ها را حفظ می‌کند

plt.imshow(medBlur)  # Display the median blurred image using matplotlib
# نمایش تصویر محوشده میانه با استفاده از matplotlib

plt.axis('off')  # Turn off axes for a cleaner display
# خاموش کردن محور‌ها برای نمایش مرتب‌تر

plt.show()  # Show the plotted median blurred image
# نمایش تصویر محوشده میانه رسم شده

```
