```python
import cv2  # OpenCV library for image and video processing
# کتابخانه OpenCV برای پردازش تصویر و ویدئو

import matplotlib.pyplot as plt  # Matplotlib library for plotting and visualizing images
# کتابخانه Matplotlib برای رسم نمودار و نمایش تصاویر

import numpy as np  # NumPy library for numerical operations and arrays
# کتابخانه NumPy برای عملیات عددی و کار با آرایه‌ها
```

```python
image = cv2.imread('./bird.jpg')  
# Read the image 'bird.jpg' from the current directory using OpenCV
# تصویر 'bird.jpg' را از پوشه فعلی با استفاده از OpenCV می‌خواند

plt.imshow(image)  
# Display the image using matplotlib
# نمایش تصویر با استفاده از matplotlib

plt.axis('off')  
# Turn off the axis to hide tick marks and labels
# خاموش کردن محور‌ها تا خطوط و برچسب‌ها نشان داده نشوند

plt.show()  
# Show the plotted image
# نمایش تصویر رسم شده
```

```python
cropped_image = image[80:280, 150:330]  
# Crop a portion of the image from row 80 to 280 and column 150 to 330
# برش بخشی از تصویر از ردیف ۸۰ تا ۲۸۰ و ستون ۱۵۰ تا ۳۳۰

plt.imshow(cropped_image)  
# Display the cropped image using matplotlib
# نمایش تصویر برش‌خورده با استفاده از matplotlib

plt.axis('off')  
# Turn off the axis to hide tick marks and labels
# خاموش کردن محور‌ها تا خطوط و برچسب‌ها نشان داده نشوند

plt.show()  
# Show the plotted cropped image
# نمایش تصویر برش‌خورده رسم شده
```

```python
resize_image = cv2.resize(image, [500, 500])  
# Resize the original image to a width and height of 500 pixels using OpenCV
# تغییر اندازه تصویر اصلی به عرض و ارتفاع ۵۰۰ پیکسل با استفاده از OpenCV

plt.imshow(resize_image)  
# Display the resized image using matplotlib
# نمایش تصویر تغییر اندازه داده شده با استفاده از matplotlib

plt.axis('off')  
# Turn off the axis to hide tick marks and labels
# خاموش کردن محور‌ها تا خطوط و برچسب‌ها نشان داده نشوند

plt.show()  
# Show the plotted resized image
# نمایش تصویر تغییر اندازه داده شده رسم شده
```

```python
resize_image = cv2.resize(image, [500, 500], interpolation=cv2.INTER_NEAREST)  
# Resize the original image to 500x500 pixels using nearest-neighbor interpolation
# تغییر اندازه تصویر اصلی به ۵۰۰x۵۰۰ پیکسل با استفاده از روش نمونه‌برداری نزدیک‌ترین همسایه

plt.imshow(resize_image)  
# Display the resized image using matplotlib
# نمایش تصویر تغییر اندازه داده شده با استفاده از matplotlib

plt.axis('off')  
# Turn off the axis to hide tick marks and labels
# خاموش کردن محور‌ها تا خطوط و برچسب‌ها نشان داده نشوند

plt.show()  
# Show the plotted resized image
# نمایش تصویر تغییر اندازه داده شده رسم شده
```

```python
resize_image = cv2.resize(image, [500, 500], interpolation=cv2.INTER_LINEAR)  
# Resize the original image to 500x500 pixels using linear interpolation
# تغییر اندازه تصویر اصلی به ۵۰۰x۵۰۰ پیکسل با استفاده از روش نمونه‌برداری خطی

plt.imshow(resize_image)  
# Display the resized image using matplotlib
# نمایش تصویر تغییر اندازه داده شده با استفاده از matplotlib

plt.axis('off')  
# Turn off the axis to hide tick marks and labels
# خاموش کردن محور‌ها تا خطوط و برچسب‌ها نشان داده نشوند

plt.show()  
# Show the plotted resized image
# نمایش تصویر تغییر اندازه داده شده رسم شده
```

```python
resize_image = cv2.resize(image, [500, 500], interpolation=cv2.INTER_CUBIC)  
# Resize the original image to 500x500 pixels using cubic interpolation (better for smooth scaling)
# تغییر اندازه تصویر اصلی به ۵۰۰x۵۰۰ پیکسل با استفاده از روش نمونه‌برداری مکعبی (برای بزرگنمایی صاف‌تر مناسب‌تر است)

plt.imshow(resize_image)  
# Display the resized image using matplotlib
# نمایش تصویر تغییر اندازه داده شده با استفاده از matplotlib

plt.axis('off')  
# Turn off the axis to hide tick marks and labels
# خاموش کردن محور‌ها تا خطوط و برچسب‌ها نشان داده نشوند

plt.show()  
# Show the plotted resized image
# نمایش تصویر تغییر اندازه داده شده رسم شده
```

```python
rotate_image = cv2.rotate(image, rotateCode=cv2.ROTATE_90_CLOCKWISE)  
# Rotate the image 90 degrees clockwise using OpenCV
# چرخش تصویر ۹۰ درجه در جهت عقربه‌های ساعت با استفاده از OpenCV

plt.imshow(rotate_image)  
# Display the rotated image using matplotlib
# نمایش تصویر چرخانده شده با استفاده از matplotlib

plt.axis('off')  
# Turn off the axis to hide tick marks and labels
# خاموش کردن محور‌ها تا خطوط و برچسب‌ها نشان داده نشوند

plt.show()  
# Show the plotted rotated image
# نمایش تصویر چرخانده شده رسم شده
```

```python
rotate_image = cv2.rotate(image, rotateCode=cv2.ROTATE_180)  
# Rotate the image 180 degrees using OpenCV
# چرخش تصویر ۱۸۰ درجه با استفاده از OpenCV

plt.imshow(rotate_image)  
# Display the rotated image using matplotlib
# نمایش تصویر چرخانده شده با استفاده از matplotlib

plt.axis('off')  
# Turn off the axis to hide tick marks and labels
# خاموش کردن محور‌ها تا خطوط و برچسب‌ها نشان داده نشوند

plt.show()  
# Show the plotted rotated image
# نمایش تصویر چرخانده شده رسم شده
```
